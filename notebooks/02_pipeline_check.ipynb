```python
import sys; sys.path.insert(0, '../src')
from features import load_and_prepare, build_preprocessor, get_train_test_split
import numpy as np

import sys
for mod in list(sys.modules.keys()):
    if 'features' in mod:
        del sys.modules[mod]
```

```python
X, y = load_and_prepare('../data/churn_raw.csv')
X_train, X_test, y_train, y_test = get_train_test_split(X, y)
```

```python
# Fit the preprocessor manually just to inspect it
preprocessor = build_preprocessor()
preprocessor.fit(X_train)
X_train_transformed = preprocessor.transform(X_train)
```

```python
print(f"Raw X_train shape:         {X_train.shape}")
print(f"Transformed X_train shape: {X_train_transformed.shape}")
```

```python
print(f"\nFirst 3 columns (numeric — should be mean≈0, std≈1):")
print(f"Mean: {X_train_transformed[:, :3].mean(axis=0).round(4)}")
print(f"Std:  {X_train_transformed[:, :3].std(axis=0).round(4)}")
```

```python
print(f"\nChurn rate train: {y_train.mean():.3f}")
print(f"Churn rate test:  {y_test.mean():.3f}")
```
